**Customers Table**

customer_id: Unique identifier for each customer

customer_unique_id: Unique identifier for each unique customer

customer_zip_code_prefix: Zip code prefix of the customer

customer_city: City of the customer

customer_state: State of the customer

**Geolocation Table**

geolocation_zip_code_prefix: Zip code prefix of the location

geolocation_lat: Latitude coordinate of the location

geolocation_lng: Longitude coordinate of the location

geolocation_city: City of the location

geolocation_state: State of the location

**Order Items Table**

order_id: Unique identifier for each order

order_item_id: Identifier for each item in the order

product_id: Unique identifier for each product

seller_id: Unique identifier for each seller

shipping_limit_date: Last date for shipping the item

price: Price of the item

freight_value: Freight value of the item

**Order Payments Table**

order_id: Unique identifier for each order

payment_sequential: Sequential number of the payment

payment_type: Type of payment

payment_installments: Number of installments for payment

payment_value: Value of the payment

**Order Reviews Table**

review_id: Unique identifier for each review

order_id: Unique identifier for each order

review_score: Score given in the review

review_comment_title: Title of the review comment

review_comment_message: Message of the review comment

review_creation_date: Creation date of the review

review_answer_timestamp: Timestamp of the review answer

**Orders Table**

order_id: Unique identifier for each order

customer_id: Unique identifier for each customer

order_status: Status of the order

order_purchase_timestamp: Timestamp of the order purchase

order_approved_at: Timestamp when the order was approved

order_delivered_carrier_date: Date when the order was delivered to the carrier

order_delivered_customer_date: Date when the order was delivered to the customer

order_estimated_delivery_date: Estimated delivery date of the order

**Products Table**

product_id: Unique identifier for each product

product_category_name: Name of the product category

product_name_lenght: Length of the product name

product_description_lenght: Length of the product description

product_photos_qty: Number of photos of the product

product_weight_g: Weight of the product in grams

product_length_cm: Length of the product in centimeters

product_height_cm: Height of the product in centimeters

product_width_cm: Width of the product in centimeters

**Sellers Table**

seller_id: Unique identifier for each seller

seller_zip_code_prefix: Zip code prefix of the seller

seller_city: City of the seller

seller_state: State of the seller

Product Category Translation Table

product_category_name: Name of the product category

product_category_name_english: Name of the product category in English

# Descriptive Statistics

**Customers Table**

**Distribution of Customer Count by Cities**

```python
city_segment = customers_df.groupby('customer_city').size().reset_index(name='customer_count')
top_cities = city_segment.sort_values(by='customer_count', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_cities['customer_city'], top_cities['customer_count'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Customer Count')
plt.title('Customer Distribution by City')
plt.xticks(rotation=45)
plt.show()
```

Analysis of the distribution of customer count by cities was conducted to determine which cities have the highest customer traffic.

**Distribution of Customer Count by States**

```python
state_segment = customers_df.groupby('customer_state').size().reset_index(name='customer_count')
plt.figure(figsize=(12, 6))
state_segment_sorted = state_segment.sort_values(by='customer_count', ascending=False)
plt.bar(state_segment_sorted['customer_state'], state_segment_sorted['customer_count'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Customer Count')
plt.title('Customer Distribution by States')
plt.xticks(rotation=45)
plt.show()
```

An analysis of the distribution of customer count by states was conducted to identify which states have a higher number of customers.

### Geolocation Table

**Order Density by Geolocation**

```python
merged_df = pd.merge(orders_df, customers_df, on='customer_id', how='inner')
order_density = merged_df['customer_zip_code_prefix'].value_counts()
order_density_df = pd.DataFrame(order_density).reset_index()
order_density_df.columns = ['customer_zip_code_prefix', 'order_density']
merged_density_df = pd.merge(order_density_df, geo_df, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
plt.figure(figsize=(10, 8))
sns.scatterplot(data=merged_density_df, x='geolocation_lng', y='geolocation_lat', size='order_density', hue='order_density', palette='YlOrRd', sizes=(20, 2000))
plt.title('Order Density by Geolocation')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Order Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Order density was examined by geographical locations to determine the regions with the highest order volume.

**Correlation between Geolocation and Delivery Times**

```python
merged_df = pd.merge(orders_df, customers_df, on='customer_id', how='left')
merged_df = pd.merge(merged_df, orderitem_df, on='order_id', how='left')
merged_df = pd.merge(merged_df, geo_df, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left')
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
merged_df['order_estimated_delivery_date'] = pd.to_datetime(merged_df['order_estimated_delivery_date'])
merged_df['delivery_time'] = (merged_df['order_estimated_delivery_date'] - merged_df['order_purchase_timestamp']).dt.days
correlation = merged_df.groupby(['geolocation_lat', 'geolocation_lng']).agg({'delivery_time': 'mean'}).reset_index()
plt.figure(figsize=(12, 6))
plt.scatter(correlation['geolocation_lng'], correlation['geolocation_lat'], c=correlation['delivery_time'], cmap='viridis', s=50, alpha=0.5)
plt.colorbar(label='Mean Delivery Time (days)')
plt.title('Correlation between Geolocation and Delivery Times')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The correlation between geographical location and delivery times was analyzed, evaluating the impact of orders from distant regions on delivery times.

**Average Coordinates by State**

```python
state_avg_coords = geo_df.groupby('geolocation_state').agg({'geolocation_lat': 'mean', 'geolocation_lng': 'mean'}).reset_index()
plt.figure(figsize=(10, 6))
plt.scatter(state_avg_coords['geolocation_lng'], state_avg_coords['geolocation_lat'], color='blue')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Average Coordinates by State')
plt.grid(True)
plt.show()
```

Average coordinates were calculated by state, determining the geographical centers of states.

**Determination of the Most Dense Cities**

```python
city_counts = geo_df['geolocation_city'].value_counts().reset_index()
city_counts.columns = ['geolocation_city', 'count']
top_cities = city_counts.head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_cities['geolocation_city'], top_cities['count'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Location Count')
plt.title('Top Cities by Location Density')
plt.xticks(rotation=45)
plt.show()
```

The most densely populated cities were identified, and the order density in these cities was examined.
